```python
import pandas as pd
```

```python
df = pd.read_csv("/content/online_retail.csv")
print(df.head(),end='\n\n')
print(df.info(),end='\n\n')
print(df.isnull().sum(),end='\n\n')
print('Types of Columns:\n',df.dtypes)
```

```python
df[df['InvoiceNo'].str.contains("C")]
```

```python
df['is_return']= df['InvoiceNo'].str.contains("C")
df['is_return'] = df['is_return'].astype(bool)
df.loc[df['is_return'], 'Quantity'] *= -1
df =df.dropna(subset=['CustomerID'])
df =df.dropna(subset=['Description'])
df['CustomerID']=df['CustomerID'].astype(int)
df.isnull().sum()
```

```python
customer = df[['CustomerID', 'Country']]
invoice = df[['CustomerID', 'InvoiceNo','InvoiceDate','is_return']]
product = df[['StockCode', 'Description','UnitPrice']]
inv_to_prod = df[['InvoiceNo', 'StockCode','Quantity']]
```

```python
customer.to_csv('customers.csv')
invoice.to_csv('invoices.csv')
product.to_csv('products.csv')
inv_to_prod.to_csv('invoice_rel_product.csv')
```
